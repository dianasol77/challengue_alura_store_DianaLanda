### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
def calcular_facturacion(data):
    # Calcular la facturación total
    data['Facturacion'] = data['Precio'] + data['Costo de envío']
    return data['Facturacion'].sum()
```

```python
# Calcular la facturación total para cada tienda
facturacion_tienda = calcular_facturacion(tienda)
facturacion_tienda2 = calcular_facturacion(tienda2)
facturacion_tienda3 = calcular_facturacion(tienda3)
facturacion_tienda4 = calcular_facturacion(tienda4)
```

```python
# Mostrar resultados
print(f'Total de facturación Tienda : {facturacion_tienda}')
print(f'Total de facturación Tienda 2: {facturacion_tienda2}')
print(f'Total de facturación Tienda 3: {facturacion_tienda3}')
print(f'Total de facturación Tienda 4: {facturacion_tienda4}')
```

```python
# Calcular la facturación por producto para cada tienda
facturacion_por_producto_tienda = tienda.groupby('Producto')['Facturacion'].sum().sort_values(ascending=False)
facturacion_por_producto_tienda2 = tienda2.groupby('Producto')['Facturacion'].sum().sort_values(ascending=False)
facturacion_por_producto_tienda3 = tienda3.groupby('Producto')['Facturacion'].sum().sort_values(ascending=False)
facturacion_por_producto_tienda4 = tienda4.groupby('Producto')['Facturacion'].sum().sort_values(ascending=False)

# Mostrar resultados por tienda
print("\nFacturación por Producto - Tienda :")
print(facturacion_por_producto_tienda)

print("\nFacturación por Producto - Tienda 2:")
print(facturacion_por_producto_tienda2)

print("\nFacturación por Producto - Tienda 3:")
print(facturacion_por_producto_tienda3)

print("\nFacturación por Producto - Tienda 4:")
print(facturacion_por_producto_tienda4)
```

# 2. Ventas por categoría

```python
# Calcular la facturación por categoría para cada tienda
facturacion_por_categoria_tienda = tienda.groupby('Categoría del Producto')['Facturacion'].sum().sort_values(ascending=False)
facturacion_por_categoria_tienda2 = tienda2.groupby('Categoría del Producto')['Facturacion'].sum().sort_values(ascending=False)
facturacion_por_categoria_tienda3 = tienda3.groupby('Categoría del Producto')['Facturacion'].sum().sort_values(ascending=False)
facturacion_por_categoria_tienda4 = tienda4.groupby('Categoría del Producto')['Facturacion'].sum().sort_values(ascending=False)
```

```python
# Mostrar resultados por tienda
print("\nFacturación por Categoría - Tienda :")
print(facturacion_por_categoria_tienda)

print("\nFacturación por Categoría - Tienda 2:")
print(facturacion_por_categoria_tienda2)

print("\nFacturación por Categoría - Tienda 3:")
print(facturacion_por_categoria_tienda3)

print("\nFacturación por Categoría - Tienda 4:")
print(facturacion_por_categoria_tienda4)
```

# 3. Calificación promedio de la tienda


```python
# Calcular la calificación promedio para cada tienda
calificacion_promedio_tienda = tienda['Calificación'].mean()
calificacion_promedio_tienda2 = tienda2['Calificación'].mean()
calificacion_promedio_tienda3 = tienda3['Calificación'].mean()
calificacion_promedio_tienda4 = tienda4['Calificación'].mean()

# Mostrar resultados
print(f"\nCalificación Promedio - Tienda : {calificacion_promedio_tienda:.2f}")
print(f"Calificación Promedio - Tienda 2: {calificacion_promedio_tienda2:.2f}")
print(f"Calificación Promedio - Tienda 3: {calificacion_promedio_tienda3:.2f}")
print(f"Calificación Promedio - Tienda 4: {calificacion_promedio_tienda4:.2f}")
```

# 4. Productos más y menos vendidos

```python
def Productos_mas_menos_vendidos(data):
    # Agrupar por producto y sumar las facturaciones
    ventas_por_producto = data.groupby('Producto')['Facturacion'].sum().sort_values(ascending=False)

    # Obtener el producto con mayor facturación
    producto_mas_vendido = ventas_por_producto.idxmax(), ventas_por_producto.max()

    # Obtener el producto con menor facturación
    producto_menos_vendido = ventas_por_producto.idxmin(), ventas_por_producto.min()

    return producto_mas_vendido, producto_menos_vendido
```

```python
# Calcular productos más y menos vendidos para cada tienda
mas_menos_tienda = Productos_mas_menos_vendidos(tienda)
mas_menos_tienda2 = Productos_mas_menos_vendidos(tienda2)
mas_menos_tienda3 = Productos_mas_menos_vendidos(tienda3)
mas_menos_tienda4 = Productos_mas_menos_vendidos(tienda4)

# Mostrar resultados
print(f"\nTienda 1 - Producto Más Vendido: {mas_menos_tienda[0][0]} (Cantidad: {mas_menos_tienda[0][1]})")
print(f"Tienda 1 - Producto Menos Vendido: {mas_menos_tienda[1][0]} (Cantidad: {mas_menos_tienda[1][1]})")

print(f"\nTienda 2 - Producto Más Vendido: {mas_menos_tienda2[0][0]} (Cantidad: {mas_menos_tienda2[0][1]})")
print(f"Tienda 2 - Producto Menos Vendido: {mas_menos_tienda2[1][0]} (Cantidad: {mas_menos_tienda2[1][1]})")

print(f"\nTienda 3 - Producto Más Vendido: {mas_menos_tienda3[0][0]} (Cantidad: {mas_menos_tienda3[0][1]})")
print(f"Tienda 3 - Producto Menos Vendido: {mas_menos_tienda3[1][0]} (Cantidad: {mas_menos_tienda3[1][1]})")

print(f"\nTienda 4 - Producto Más Vendido: {mas_menos_tienda4[0][0]} (Cantidad: {mas_menos_tienda4[0][1]})")
print(f"Tienda 4 - Producto Menos Vendido: {mas_menos_tienda4[1][0]} (Cantidad: {mas_menos_tienda4[1][1]})")
```

# 5. Envío promedio por tienda

```python
def envio_promedio(data):
    # Calcular el costo de envío promedio
    promedio_envio = data['Costo de envío'].mean()
    return promedio_envio
```

```python
# Calcular el envío promedio para cada tienda
envio_promedio_tienda = envio_promedio(tienda)
envio_promedio_tienda2 = envio_promedio(tienda2)
envio_promedio_tienda3 = envio_promedio(tienda3)
envio_promedio_tienda4 = envio_promedio(tienda4)

# Mostrar resultados
print(f"\nEnvío Promedio - Tienda 1: {envio_promedio_tienda:.2f}")
print(f"Envío Promedio - Tienda 2: {envio_promedio_tienda2:.2f}")
print(f"Envío Promedio - Tienda 3: {envio_promedio_tienda3:.2f}")
print(f"Envío Promedio - Tienda 4: {envio_promedio_tienda4:.2f}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Gráfico 1: Ingresos por tienda
tiendas = ['Tienda ', 'Tienda 2', 'Tienda 3', 'Tienda 4']
ingresos = [tienda['Facturacion'].sum(), tienda2['Facturacion'].sum(), tienda3['Facturacion'].sum(), tienda4['Facturacion'].sum()]
```

```python
plt.figure(figsize=(10, 6))
plt.bar(tiendas, ingresos, color='skyblue')
plt.title('Ingresos por Tienda')
plt.xlabel('Tienda')
plt.ylabel('Ingresos')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

```python
# Agregar una columna para identificar la tienda
tienda['Tienda'] = 'Tienda'
tienda2['Tienda'] = 'Tienda 2'
tienda3['Tienda'] = 'Tienda 3'
tienda4['Tienda'] = 'Tienda 4'

# Concatenar todos los DataFrames
datos_tiendas = pd.concat([tienda, tienda2, tienda3, tienda4], ignore_index=True)

# Gráfico 1: Ingresos Totales por Tienda
ingresos_totales = datos_tiendas.groupby('Tienda')['Facturacion'].sum().reset_index()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=ingresos_totales, x='Tienda', y='Facturacion', palette='viridis')
plt.title('Ingresos Totales por Tienda')
plt.xlabel('Tienda')
plt.ylabel('Ingresos')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

```python
# Gráfico de Cajas: Distribución de Cantidades Vendidas por Tienda (usando Facturación como proxy de valor de venta)
plt.figure(figsize=(10, 6))
sns.boxplot(data=datos_tiendas, x='Tienda', y='Facturacion', palette='Set3')
plt.title('Distribución de Facturación por Tienda')
plt.xlabel('Tienda')
plt.ylabel('Facturación (Monto de Venta)')
plt.grid(axis='y')
plt.show()
```
